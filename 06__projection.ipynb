# Projection

```python
import numpy as np
import plotvec as plotvec
import matplotlib.pyplot as plt
```

```python
a = np.array([1, 2])
b = np.array([2, 1])
plotvec.plotvec(a, b)
plt.annotate('a', (a[0], a[1] - 0.25))
plt.annotate('b', (b[0], b[1] - 0.25))
```

```python
# angle between two vectors
a = np.array([1, 2])
b = np.array([2, 1])
# cos(theta) = (a dot b) / (||a|| ||b||)
# norm: length of vector ||a|| = sqrt(a1^2 + a2^2 + ... + an^2)
cos_theta = np.dot(a, b) /( np.sqrt(np.dot(a, a)) * np.sqrt(np.dot(b, b)))
# np.across(): inverse cosine, returns angle in radians
# radian -> degree : degree = radian * 180 / pi
theta = np.arccos(cos_theta)
print(round(theta * 180 / np.pi), 2)
```

```python

# find unit vector
a_t = a / np.linalg.norm(a) # linalg.norm(): return length of vector
plotvec.plotvec(a, b)
plt.annotate('a', (a[0], a[1] - 0.25))
plt.annotate('b', (b[0], b[1] - 0.25))
plotvec.plotvec(a_t, newfig=False, color_offset=2, width=0.015)
plt.annotate('unit $a$', (a_t[0], a_t[1] - 0.25))
```

```python

# projection of vector a onto vector b
# 먼저, scalar projection을 구한다. 즉, a를 b 방향으로 얼마나 뻗었는지 구한다.
# scalar_proj = (a dot b) / ||b|| = |a| cos(theta)
scalar_proj = np.dot(a, b) / np.linalg.norm(b)

# vector projection: scalar_proj * (b / ||b||) = scalar_proj * unit_b
# 즉, scalar projection에 b의 단위벡터를 곱해준다.
vector_proj = scalar_proj * (b / np.linalg.norm(b))
plotvec.plotvec(a, b)
plt.annotate('a', (a[0], a[1] - 0.25))
plt.annotate('b', (b[0], b[1] - 0.25))
plotvec.plotvec(a_t, newfig=False, color_offset=2, width=0.015)
plt.annotate('unit $a$', (a_t[0], a_t[1] - 0.25))
plotvec.plotvec(vector_proj, newfig=False, color_offset=3, width=0.02)
plt.annotate('proj $a$ to $b$', (vector_proj[0], vector_proj[1] - 0.25))
```

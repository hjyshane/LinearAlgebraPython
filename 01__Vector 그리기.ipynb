```python
# Import library

import numpy as np
import matplotlib.pyplot as plt
from math import acos, degrees
```

```python
# define sample vectors v, w
v=np.array([[1,2,3,4,5]])
w=np.array([[9,8,7,6,5]])
```

```python
# compute v + w, 2*v, -3*w
vplusw=v+w
twov=2 * v
minthreew= -3 * w
```

## Plotting vetrors
quiver([X, Y], U, V, [C], /, **kwargs)

plt.quiver 함수는 벡터를 화살표로 그려주는 함수이다.

X, Y는 화살표의 시작 위치, U, V는 화살표의 방향, C는 선택적으로 색상을 설정한다. X, Y, U, V, C는 위치 전용 인수이다.

x,y는 벡터의 시작 위치를 나타내는 배열이다. 여기서는 원점에서 시작하므로 모두 0으로 설정한다.

U,V는 각각 벡터의 x성분과 y성분을 나타내는 배열이다.

여러 벡터를 그릴때는 x,y,U,V 모두 배열로 설정해야 한다.

예를 들어, v벡터의 x성분은 v[0], y성분은 v[1]이다. 그 열에 있늰 모든 벡터들이 한번에 그려진다.

scale=1로 설정하면 벡터의 크기가 그대로 그려진다.

angles='xy', scale_units='xy' 옵션은 벡터의 방향을 좌표축과 동일하게 맞춰준다.

```python
# Set origin
x = np.zeros(5) # 원점 좌표
y = np.zeros(5) # 원점 좌표
    
# 하나의 벡터만 보여줄때
plt.plot() # 새로운 그래프 생성
plt.quiver(0,0,w[0,0],w[0,0], angles='xy', scale_units='xy', scale=1, color='r', label='w')
plt.xlim(0,20) # x축 범위 설정
plt.ylim(0,20) # y축 범위 설정
plt.grid() # 격자무늬 표시
plt.show() # 그래프 출력

# 열벡터 여러개를 한번에 보여줄때
plt.plot()
plt.quiver(x,y,w[0],w[0], angles='xy', scale_units='xy', scale=1, color='r')
plt.quiver(x,y,v[0],v[0], angles='xy', scale_units='xy', scale=1, color='b')
plt.quiver(x,y,vplusw[0],vplusw[0], angles='xy', scale_units='xy', scale=1, color='g')
plt.xlim(0,20)
plt.ylim(0,20)
plt.grid()
plt.show()

```

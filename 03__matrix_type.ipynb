# Transpose

```python
import numpy as np
import sympy as sp


a,b,c,d,i,j,k,h = sp.symbols('a b c d i j k h')
```

# 1. Transpose (.T)
대각선을 기준으로 회전

```python
m = sp.Matrix([
    ["a", "b", "c"],
    ["d", "e", "f"],
    ["g", "h", "i"]
            ])

m.T
```

```python
print(np.transpose(m))  # 같은 결과
```

# 2. Inverse (np.linalg.inv())

```python
A = np.array([[3, 0, 2], [2, 0, -2], [0, 1, 1]])
A_inv = np.linalg.inv(A)
A_inv
```

```python
m = sp.Matrix([
    ["a", "b", "c"],
    ["d", "e", "f"],
    ["g", "h", "i"]
            ])
sp.Inverse(m)
```
